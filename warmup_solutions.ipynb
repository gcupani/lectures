````python
from astropy.io import fits
import matplotlib.pyplot as plt
import numpy as np
from utils import plot_graph,plot_img

name = 'sci_2015/Autosave Image -001.fit'
hdul = fits.open(name)
img = hdul[0].data
sub_img = img[510:530, 630:660]
````

> **Your turn now**: write the code to extract a 11x11-pixel window centered in (7, 13) from ```sub_img``` and to plot all the rows and columns of this new sub image.

````python
sub_img_2 = sub_img[7-5:7+6,13-5:13+6]
plot_img(sub_img_2)
for i in range(11):
    r = sub_img_2[i,:]
    c = sub_img_2[:,i]
    plot_graph(range(len(r)), r, xlabel='pixel', ylabel='ADU', new=i==0)
    plot_graph(range(len(c)), c, new=False, mode='scatter')
plt.show()
````

> **Only for the brave**: using Python lists, write a function to swap the x and y axes of an image. **Tip**: to shorten the computation time, try swapping ```sub_img``` instead of ```img```.

````python
sub_img_swap = []
for j in range(sub_img.shape[1]):
    sub_img_col = []
    for i in range(sub_img.shape[0]):
        sub_img_col.append(sub_img[i,j])
    sub_img_swap.append(sub_img_col) 
plot_img(sub_img)
plot_img(sub_img_swap)
````

> **Your turn now**: with help from the NumPy Reference (https://docs.scipy.org/doc/numpy-1.16.1/reference/), compute the arithmetic mean and the median of ```sub_img```. What can we infer from the difference between the two values? **Extra points**: compute the median of each rows of ```img``` separately and plot all the values as a function of the row index.

````python
print("Mean: %3.2f, median: %3.2f" % (np.mean(sub_img), np.median(sub_img)))

row_med = []
for row in img:
    row_med.append(np.median(row))
    
row_med_vect = np.median(img, axis=1)
    
plot_graph(range(len(row_med)), row_med, xlabel='y pixel', ylabel='median ADU', linewidth=4)
plot_graph(range(len(row_med_vect)), row_med_vect, new=False, linewidth=1)

plt.show()
````
